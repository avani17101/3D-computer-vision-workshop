# Simple Perceptron for 1D dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

A1 = np.array([-1,1])
A2 = np.array([2,1])
B1 = np.array([3,1])
B2 = np.array([6,1])
w = 5*np.random.rand(2)

cost = max(0,-np.dot(w,A1)) + max(0,-np.dot(w,A2)) + max(0,np.dot(w,B1)) + max(0,np.dot(w,B2))

itr = 1

while (cost > 0):
    if -np.dot(w,A1) > 0:
        w = w + A1
    if -np.dot(w,A2) > 0:
        w = w + A2
    if np.dot(w,B1) > 0:
        w = w - B1
    if np.dot(w,B2) > 0:
        w = w - B2
    cost = max(0,-np.dot(w,A1)) + max(0,-np.dot(w,A2)) + max(0,np.dot(w,B1)) + max(0,np.dot(w,B2))
    
    c = -w[1]/w[0]
    plt.scatter([c], [1], c='black', label='Classifier')
    plt.scatter([A1[0],A2[0]],[A1[1],A2[1]],c='red')
    plt.scatter([B1[0],B2[0]],[B1[1],B2[1]],c='blue')
    plt.title('Iteration '+str(itr))
    plt.xlabel('x', color='#1C2833')
    plt.ylabel('y', color='#1C2833')
    plt.legend(loc='upper left')
    plt.grid()
    plt.show()
    
    itr += 1
```

```python
def z_function(x, y):
    z = [[0 for i in range(0,100)] for j in range(0,100)]
    for i in range(0,100):
        for j in range(0,100):
            w = np.array([x[i][j],y[i][j]])
            z[i][j] = max(0,-np.dot(w,A1)) + max(0,-np.dot(w,A2)) + max(0,np.dot(w,B1)) + max(0,np.dot(w,B2))
    return np.array(z)
```

```python
def z_function_2(x, y):
    z = [[0 for i in range(0,100)] for j in range(0,100)]
    for i in range(0,100):
        for j in range(0,100):
            w = np.array([x[i][j],y[i][j]])
            z[i][j] = cnt(-np.dot(w,A1)) + cnt(-np.dot(w,A2)) + cnt(np.dot(w,B1)) + cnt(np.dot(w,B2))
    return np.array(z)
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")

x = np.linspace(-2000, 2000, 100)
y = np.linspace(-2000, 2000, 100)
X, Y = np.meshgrid(x, y)
Z = z_function(X, Y)

ax.plot_wireframe(X, Y, Z, color='green')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()

x = np.linspace(-200, 200, 100)
y = np.linspace(-200, 200, 100)
X, Y = np.meshgrid(x, y)
Z = z_function_2(X, Y)

plt.scatter(X, Y, Z, color='green')

plt.show()
```
